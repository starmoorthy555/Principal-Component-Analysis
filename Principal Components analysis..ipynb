```python
#Importing library.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating a random numbers.
rng = np.random.RandomState(1)
x = np.dot(rng.rand(2, 2),rng.randn(2, 200)).T
```

```python
#plot the values
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
#importing PCA Library.
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
```

```python
pca.fit(x)
```

```python
y = pca.transform(x)
```

```python
y_new = pca.inverse_transform(y)
```

```python
#Visualization the PCA value.
plt.scatter(x[:,0],x[:,1],alpha=0.4)
plt.scatter(y_new[:,0],y_new[:,1],alpha=0.8)
plt.show()
```
